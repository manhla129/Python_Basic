```python
# 1. Số phức
import random
class Sophuc:
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def print(self):
        vt1 = ''
        vt2 = ''
        if self.a != 0:
            vt1 = str(self.a)
        if self.b == 1:
            if self.a == 0:
                vt2 = 'i'
            else:
                vt2 = ' + i'
        elif self.b == -1:
            if self.a == 0:
                vt2 = '-i'
            else:
                vt2 = ' - i'
        elif self.b > 0:
            if self.a == 0:
                vt2 = str(self.b) + 'i'
            else:
                vt2 = ' + ' + str(self.b) + 'i'
        elif self.b < 0:
            if self.a == 0:
                vt2 = str(self.b) + 'i'
            else:
                vt2 = ' - ' + str(-self.b) + 'i'
        return '(' + vt1 + vt2 + ')'
    
    def module(self):
        return (self.a**2 + self.b**2)**0.5
    
list_sophuc = []
list_module = []
i = 0
for i in range(100):
        temp = Sophuc(random.randint(-10,10),random.randint(-10,10))
        list_sophuc.append(temp.print())
        if i < 10:
            print(temp.print())
        list_module.append(temp.module())

max = list_module[0]
min = list_module[1]
max_index = 0
min_index = 1

for i in range(len(list_module)):
    if list_module[i] > max:
        max = list_module[i]
        max_index = i
    if list_module[i] < min:
        min = list_module[i]
        min_index = i
        
print('Max module:', list_sophuc[max_index],list_module[max_index])
print('Min module:', list_sophuc[min_index],list_module[min_index])
```

```python
# 2. Phân số
import random
class Phanso:
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def print(self):
        vt1 = ''
        vt2 = ''
        if self.a%self.b == 0:
            self.a = self.a/self.b
            self.b = 1
        if self.b == 0:
            return '#N/A'
        elif self.a == 0:
            return '0'
        elif self.b == 1:
            return str(self.a)
        elif self.b == -1:
            return str(-self.a)
        else:
            if self.b < 0:
                vt1 = str(-self.a)
                vt2 = str(-self.b)
            else:
                vt1 = str(self.a)
                vt2 = str(self.b)
            return vt1 + '/' + vt2
    
    def sosanh(self, other):
        if (self.a/self.b) > (other.a/other.b):
            return self.print() + ' > ' + other.print()
        elif (self.a/self.b) == (other.a/other.b):
            return self.print() + ' =' + other.print()
        else:
            return self.print() + ' < ' + other.print()
        
    def tong(self, other):
        return Phanso((self.a*other.b + other.a*self.b), (self.b*other.b)).print()
    
    def hieu(self, other):
        return Phanso((self.a*other.b - other.a*self.b), (self.b*other.b)).print()
    
    def tich(self, other):
        return Phanso((self.a*other.a), (self.b*other.b)).print()
    
    def thuong(self, other):
        return Phanso((self.a/other.a), (self.b/other.b)).print()
    
    def value(self):
        return self.a/self.b
    
ps1 = Phanso(3,-4)
print('PS1:',ps1.print())
ps2 = Phanso(14,7)
print('PS2:',ps2.print())
print('So sánh:', ps1.sosanh(ps2))
print('Tổng:', ps1.tong(ps2))
print('Hiệu:', ps1.hieu(ps2))
print('Tích:', ps1.tich(ps2))
print('Thương:', ps1.thuong(ps2))
```
