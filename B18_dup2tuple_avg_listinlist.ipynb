```python
#Tạo tuple1 có n phần tử. 
#Từ tuple1 tạo ra tuple2 cũng có n phần tử với mỗi phần tử ở vị trí tương ứng có giá trị x2 ở tuple1
#Tìm ra các phần tử xuất hiện ở cả 2 tuple

import random
n = int(input('Nhập n: ')) 

list1 = [] 
for i in range(n):
    list1.append(random.choice([i for i in range(10)]))
print('\tTuple1: ', tuple(list1))

list2 = [] 
for i in list1:
    list2.append(i*2)
print('\tTuple2: ', tuple(list2))

list_dup = []
for i in list1:
    for j in list2:
        if i == j:
            cond = True
            for k in list_dup:
                if k == i:
                    cond = False
                    break
            if cond:
                list_dup.append(i)
list_dup.sort()
print('\tList dup: ',list_dup)

```

```python
#Tạo list A gồm m phần tử, mỗi phần tử của list A là 1 list khác gồm n phần tử là các số tự nhiên
#Tính giá trị trung bình mỗi phần tử trong list A
import random
m = int(input('Nhập m: ')) 
n = int(input('Nhập n: ')) 

A = []
for i in range(m):
    temp = []
    for j in range(n):
        temp.append(random.choice([i for i in range(10)]))
    A.append(temp)
print('List A: ', A)

A_tb = []
for i in A:
    sum = 0
    for j in i:
        sum += j
    A_tb.append(sum/len(i))
print('List A_tb: ',A_tb)
```
