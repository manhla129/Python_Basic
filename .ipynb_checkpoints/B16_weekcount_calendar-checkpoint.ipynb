```python
import datetime
from calendar import monthrange

def weekcount(month, year):
    '''Hàm đếm số tuần lẻ, chẵn của tháng
    Đầu vào lần lượt là: tháng, năm'''
    firstdate = datetime.date(year, month, 1)
    num_day = monthrange(year,month)[1]
    list_sun =''
    count = 0
    week_count=0
    for i in range(1,num_day+1):
        if datetime.date(year, month, i).weekday() == 6:
            list_sun+=str(i)+' '
            count +=1
    if firstdate.weekday() == 6:
        week_count = count
    else:
        week_count = count +1
    tuanle=0
    if list_sun[0] !='7':
        tuanle +=1
    if list_sun[-1] != str(num_day):
        tuanle +=1
    tuanchan=week_count-tuanle
    return tuanle, tuanchan

##############################

weekcount(4,2022)
```

```python
import turtle
import datetime
from calendar import monthrange

def calendar(month, year):
    '''Hàm vẽ lịch theo tháng, năm nhập vào
    Đầu vào lần lượt là: tháng, năm
    '''
    #Khai báo biến
    tl = turtle.Turtle()
    tl.speed(30)
    const = 40
    h_text = int(const/10)
    monthname = ['Jan','Feb','Mar','Apr','May','Jun','July','Aug','Sep','Oct','Nov','Dec']
    weekdayname = ['Mo','Tu','We','Th','Fr','Sa','Su']
    tl.penup()
    tl.goto(-200,200)
    tl.forward(h_text)
    tl.pendown()
    
    # tên tháng
    tl.left(90)
    tl.pencolor('green')
    tl.write(' '*7 + monthname[month-1], font=('Arial',60,'bold'))
    tl.penup()
    tl.backward(const*1.2)
    tl.pendown()
    tl.pencolor('red')
    tl.write(' '*18 + str(year), font = ('Arial',30,'italic'))
    tl.pencolor('black')
    tl.penup()
    tl.backward(h_text)
    tl.backward(const)
    tl.pendown()
    
    # vẽ khung thứ, tên thứ
    tl.forward(h_text)
    for i in range(7):
        if i==5:
            tl.pencolor('blue')
        elif i==6:
            tl.pencolor('red')
        tl.write(' '*int(h_text/2) + weekdayname[i], font = ('Arial',12,'bold'))
        tl.penup()
        tl.right(90)
        tl.forward(const)
        tl.left(90)
        tl.pendown()
    tl.pencolor('black')
    tl.backward(h_text)
    tl.right(90)
    tl.backward(const*7)
    tl.penup()
    tl.left(90)
    tl.backward(const)
    tl.right(90)
    tl.pendown()
    
    # điền ngày
    first_day = datetime.date(year,month,1).weekday()
    max_day = monthrange(year,month)[1] - 1
    list_day =''
    for i in range(42):
        if i < first_day or i > max_day + first_day:
            list_day+='-,'
        else:
            list_day+=(str(i-first_day+1) if i-first_day+1>9 else '  '+ str(i-first_day+1)) + ','
    i=1
    list_day=list_day.split(',')
    while i <= 42:
        if (i+1)%7==0:
            tl.pencolor('blue')
        elif i%7==0:
            tl.pencolor('red')
        tl.write(' '*int(h_text/2) + list_day[i-1] if list_day[i-1] !='-' else ' ', font=('Arial',12,'normal'))
        if i%7==0:
            tl.penup()
            tl.right(90)
            tl.forward(const)
            tl.left(90)
            tl.backward(const*6)
            tl.pendown()
        else:
            tl.penup()
            tl.forward(const)
            tl.pendown()
        i+=1
        tl.pencolor('black')
    
    # vẽ khung
    tl.backward(const)
    tl.forward(const*9)
    tl.left(90)
    tl.forward(const*12)
    tl.left(90)
    tl.forward(const*9)
    tl.left(90)
    tl.forward(const*12)
    
    tl.penup()
    tl.forward(const/10)
    tl.left(90)
    tl.backward(const/10)
    tl.pendown()
    
    tl.pensize(2)
    tl.forward(const*9 + const/5)
    tl.left(90)
    tl.forward(const*12 + const/5)
    tl.left(90)
    tl.forward(const*9 + const/5)
    tl.left(90)
    tl.forward(const*12 + const/5)
    tl.hideturtle()
    turtle.done()
    
####################################

calendar(4,2022)
```
