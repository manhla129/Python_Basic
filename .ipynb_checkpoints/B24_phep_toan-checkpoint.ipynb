```python
from tkinter import *

def cong(event):
    try:
        num1 = int(entryInput1.get())
        num2 = int(entryInput2.get())
        labelOutput['text'] = str(num1) + ' + ' + str(num2) + ' = ' + str(num1+num2)
        labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
    
def tru(event):
    try:
        num1 = int(entryInput1.get())
        num2 = int(entryInput2.get())
        labelOutput['text'] = str(num1) + ' - ' + str(num2) + ' = ' + str(num1-num2)
        labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
def nhan(event):
    try:
        num1 = int(entryInput1.get())
        num2 = int(entryInput2.get())
        labelOutput['text'] = str(num1) + ' * ' + str(num2) + ' = ' + str(num1*num2)
        labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
def chia(event):
    try:
        num1 = int(entryInput1.get())
        num2 = int(entryInput2.get())
        if num2 ==  0:
            labelOutput['text'] = 'Lỗi chia cho 0'
        else:
            labelOutput['text'] = str(num1) + ' / ' + str(num2) + ' = ' + str(num1/num2)
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
        
window = Tk()
window.geometry('400x250')
window.title('Toán tử')

titleInput1 = Label(window, text='Num 1:')
titleInput1.place(x=10, y= 10)
titleInput2 = Label(window, text='Num 2:')
titleInput2.place(x=10, y= 50)
titleOutput = Label(window, text='Output:')
titleOutput.place(x=10, y= 90)

entryInput1 = Entry(window, text = '', width = 35)
entryInput1.place(x=100, y = 10)
entryInput2 = Entry(window, text = '', width = 35)
entryInput2.place(x=100, y = 50)
labelOutput = Label(window, text = '', borderwidth = 1)
labelOutput.place(x=100, y = 90)
entryInput1.focus()

btcong = Button(window, text='+')
btcong.bind('<Button-1>', cong)
btcong.place(x=80, y = 140, width=40, height=30)

bttru = Button(window, text='-')
bttru.bind('<Button-1>', tru)
bttru.place(x=140, y = 140, width=40, height=30)

btnhan = Button(window, text='*')
btnhan.bind('<Button-1>', nhan)
btnhan.place(x=200, y = 140, width=40, height=30)

btchia = Button(window, text='/')
btchia.bind('<Button-1>', chia)
btchia.place(x=260, y = 140, width=40, height=30)

window.mainloop()
```
