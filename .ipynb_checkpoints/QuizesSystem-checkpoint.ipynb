```python
class Answer:
    def __init__(self, text, iscorrect):
        self.text = text
        self.iscorrect = iscorrect
        
    def display(self):
        print("Đáp án:", self.text, end=' ')
        print(self.iscorrect)
        
    def clone(self):
        return Answer(self.text, False)
        
class Question:
    def __init__(self, text, ans):
        self.text = text
        self.ans = ans
    
    def display(self):
        print("Nội dung câu hỏi:", self.text)
        for i in self.ans:
            i.display()
            
    def clone(self):
        ans_copy = []
        for i in self.ans:
            ans_copy.append(i.clone())
        return Question(self.text, ans_copy)
    
    def not_answer(self):
        r = True
        for i in self.ans:
            if i.iscorrect == True:
                r = False
                break
        return r

    def correct_answer(self, q):
        r = True
        for i in range(len(self.ans)):
            if self.ans[i].iscorrect != q.ans[i].iscorrect:
                r = False
                break
        return r
                
            
class Test:
    def __init__(self, test):
        self.test = test
    
    def clone(self):
        copy = []
        for i in self.test:
            copy.append(i.clone())
        return Test(copy)
    
    def compare(self, t):
        answered = 0
        correct_answer = 0
        not_answer = 0
        total = len(self.test)
        for i in range(len(self.test)):
            if self.test[i].not_answer():
                not_answer += 1
            if self.test[i].correct_answer(t.test[i]):
                correct_answer += 1
        return total, total - answered, correct_answer
    
    def loadQuestion(self, path):
        f = open('data.txt','r', encoding='utf8')
        for line in f:
            items = line.split('|')
            print(items[5])
            print(bool(items[5]))
            ans1 = Answer(items[1], True if items[5]=='True' else False)
            ans2 = Answer(items[2],True if items[6]=='True' else False)
            ans3 = Answer(items[3],True if items[7]=='True' else False)
            ans4 = Answer(items[4],True if items[8]=='True' else False)
            ans = [ans1, ans2, ans3, ans3]
            q = Question(items[0], ans)
            self.test.append(q)
            
             
    def createThreeQuestion(self):
        ans1 = Answer('Việt Nam',True)
        ans2 = Answer('Mexico',False)
        ans3 = Answer('Pháp',False)
        ans4 = Answer('Úc',False)
        ans01 = [ans1, ans2, ans3, ans4]         
        q1 = Question('Nước nào thuộc Châu Á?', ans01)

        ans5 = Answer('Mexico',False)
        ans6 = Answer('Việt Nam',True)
        ans7 = Answer('Pháp',False)
        ans8 = Answer('Úc',False)
        ans02 = [ans5, ans6, ans7, ans8]         
        q2 = Question('Hà Nội là thủ đô nước nào?', ans02)

        ans9 = Answer('Mexico',False)
        ans10 = Answer('Việt Nam',False)
        ans11 = Answer('Trung Quốc',False)
        ans12 = Answer('Nga',True)
        ans03 = [ans9, ans10, ans11, ans12]         
        q3 = Question('Nước nào rộng nhất thế giới?', ans03)
        self.test = [q1, q2, q3]
      
```

```python
from tkinter import *

index = 0
finished = False
lq = []
original = Test(lq)
# original.createThreeQuestion()
original.loadQuestion('data.txt')

t = original.clone()

def displayQuestion(i):
    question_text.delete(1.0,'end')
    question_text.insert(1.0,t.test[i].text)
    
    text_answer1.delete(1.0,'end')
    text_answer1.insert(1.0,t.test[i].ans[0].text)
    cb1.set(t.test[i].ans[0].iscorrect)
    
    text_answer2.delete(1.0,'end')
    text_answer2.insert(1.0,t.test[i].ans[1].text)
    cb2.set(t.test[i].ans[1].iscorrect)
    
    text_answer3.delete(1.0,'end')
    text_answer3.insert(1.0,t.test[i].ans[2].text)
    cb3.set(t.test[i].ans[2].iscorrect)
    
    text_answer4.delete(1.0,'end')
    text_answer4.insert(1.0,t.test[i].ans[3].text)
    cb4.set(t.test[i].ans[3].iscorrect)
    
    global finished
    if finished == True:
        ckbtn1['bg']='white'
        ckbtn1['state'] = 'disable'
        ckbtn2['bg']='white'
        ckbtn2['state'] = 'disable'
        ckbtn3['bg']='white'
        ckbtn3['state'] = 'disable'
        ckbtn4['bg']='white'
        ckbtn4['state'] = 'disable'
        
        if t.test[i].correct_answer(original.test[i]):
            for j in range(len(t.test[i].ans)):
                if t.test[i].ans[j].iscorrect:
                    if j == 0: 
                        ckbtn1['bg']='green'
                        text_answer1['bg']='green'
                    if j == 1: 
                        ckbtn2['bg']='green'
                        text_answer2['bg']='green'
                    if j == 2: 
                        ckbtn3['bg']='green'
                        text_answer3['bg']='green'
                    if j == 3: 
                        ckbtn4['bg']='green'
                        text_answer4['bg']='green'
        else:
            for j in range(len(t.test[i].ans)):
                if t.test[i].ans[j].iscorrect:
                    if j == 0: 
                        ckbtn1['bg']='red'
                        text_answer1['bg']='red'
                    if j == 1: 
                        ckbtn2['bg']='red'
                        text_answer2['bg']='red'
                    if j == 2: 
                        ckbtn3['bg']='red'
                        text_answer3['bg']='red'
                    if j == 3: 
                        ckbtn4['bg']='red'
                        text_answer4['bg']='red'
    
def goto_nextquestion(event):
    global index
    if index < len(t.test)-1:
        index +=1
        btn_previous['state'] = 'active'
        if index == len(t.test)-1:
             btn_next['state'] = 'disable'
    displayQuestion(index)
    
def goto_previousquestion(event):
    global index    
    if index > 0:
        index -=1
        btn_next['state'] = 'active'
        if index == 0:
             btn_previous['state'] = 'disable'
    displayQuestion(index)
    
def get_temporary_ans1():
    global index
    t.test[index].ans[0].iscorrect = cb1.get()
    
def get_temporary_ans2():
    global index
    t.test[index].ans[1].iscorrect = cb2.get()
    
def get_temporary_ans3():
    global index
    t.test[index].ans[2].iscorrect = cb3.get()
    
def get_temporary_ans4():
    global index
    t.test[index].ans[3].iscorrect = cb4.get()
    
def completed_test(event):
    global finished
    finished = True
    total, answered, correct_answer = t.compare(original)
    lbResult['text'] = f'Tổng số: {total}, trả lời: {answered}, đúng: {correct_answer}'
    
root = Tk()
root.geometry('750x600')
root.resizable(0,0)
root.title('Question')

lbFrame_question = LabelFrame(root, text='Nội dung câu hỏi',font=('Helvetica',12), fg='blue')
lbFrame_question.pack(expand=True, fill=BOTH)
question_text = Text(lbFrame_question, font=('Helvetica',14), height=5)
question_text.pack(expand=True, fill=BOTH, padx=10, pady=10)


lbFrame_answers = LabelFrame(root, text='Các đáp án',font=('Helvetica',12), fg='blue')
lbFrame_answers.pack(expand=True, fill=BOTH)
cb1 = BooleanVar()
cb1.set(0)
cb2 = BooleanVar()
cb2.set(0)
cb3 = BooleanVar()
cb3.set(0)
cb4 = BooleanVar()
cb4.set(0)

ckbtn1 = Checkbutton(lbFrame_answers, variable=cb1, text='', onvalue=True, offvalue=False, 
                     command=get_temporary_ans1)
ckbtn1.grid(row=0, column=0, padx=5, pady=5)
ckbtn2 = Checkbutton(lbFrame_answers, variable=cb2, text='', onvalue=True, offvalue=False, 
                     command=get_temporary_ans2)
ckbtn2.grid(row=0, column=2, padx=5, pady=5)
ckbtn3 = Checkbutton(lbFrame_answers, variable=cb3, text='', onvalue=True, offvalue=False, 
                     command=get_temporary_ans3)
ckbtn3.grid(row=1, column=0, padx=5, pady=5)
ckbtn4 = Checkbutton(lbFrame_answers, variable=cb4, text='', onvalue=True, offvalue=False, 
                     command=get_temporary_ans4)
ckbtn4.grid(row=1, column=2, padx=5, pady=5)

text_answer1 = Text(lbFrame_answers, font=('Helvetica',12), height=5, width=35)
text_answer1.grid(row=0, column=1, padx=5, pady=5)
text_answer2 = Text(lbFrame_answers, font=('Helvetica',12), height=5, width=35)
text_answer2.grid(row=0, column=3, padx=5, pady=5)
text_answer3 = Text(lbFrame_answers, font=('Helvetica',12), height=5, width=35)
text_answer3.grid(row=1, column=1, padx=5, pady=5)
text_answer4 = Text(lbFrame_answers, font=('Helvetica',12), height=5, width=35)
text_answer4.grid(row=1, column=3, padx=5, pady=5)

lbFrame_action = LabelFrame(root, text='Thao tác',font=('Helvetica',14), fg='blue')
lbFrame_action.pack(expand=True, fill=BOTH)

photo_previous = PhotoImage(file=r'IMG\previous.png')
btn_previous = Button(lbFrame_action, text='Câu hỏi trước',font=('Helvetica',8),
                     image=photo_previous, compound=LEFT)
btn_previous.place(x=10, y=10)
btn_previous.bind('<Button-1>', goto_previousquestion)

photo_next = PhotoImage(file=r'IMG\next.png')
btn_next = Button(lbFrame_action, text='Câu hỏi tiếp theo',font=('Helvetica',8),
                     image=photo_next, compound=RIGHT)
btn_next.place(x=140, y=10)
btn_next.bind('<Button-1>', goto_nextquestion)

photo_check = PhotoImage(file=r'IMG\check.png')
btn_end = Button(lbFrame_action, text='Kết thúc',font=('Helvetica',8), fg='red',
                     image=photo_check, compound=LEFT)
btn_end.place(x=400, y=10)
btn_end.bind('<Button-1>', completed_test)

lbResult = Label(lbFrame_action, text='')
lbResult.place(x=550, y=10)

displayQuestion(0)
btn_previous['state'] = 'disable'


root.mainloop()
```
