```python
# 1. Số phức
import random
class Sophuc:
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def print(self):
        vt1 = ''
        vt2 = ''
        if self.a != 0:
            vt1 = str(self.a)
        if self.b == 1:
            if self.a == 0:
                vt2 = 'i'
            else:
                vt2 = ' + i'
        elif self.b == -1:
            if self.a == 0:
                vt2 = '-i'
            else:
                vt2 = ' - i'
        elif self.b > 0:
            if self.a == 0:
                vt2 = str(self.b) + 'i'
            else:
                vt2 = ' + ' + str(self.b) + 'i'
        elif self.b < 0:
            if self.a == 0:
                vt2 = str(self.b) + 'i'
            else:
                vt2 = ' - ' + str(-self.b) + 'i'
        return '(' + vt1 + vt2 + ')'
    
    def module(self):
        return (self.a**2 + self.b**2)**0.5
    
list_sophuc = []
list_module = []
i = 0
for i in range(100):
        temp = Sophuc(random.randint(-10,10),random.randint(-10,10))
        list_sophuc.append(temp.print())
        if i < 10:
            print(temp.print())
        list_module.append(temp.module())

max = list_module[0]
min = list_module[1]
max_index = 0
min_index = 1

for i in range(len(list_module)):
    if list_module[i] > max:
        max = list_module[i]
        max_index = i
    if list_module[i] < min:
        min = list_module[i]
        min_index = i
        
print('Max module:', list_sophuc[max_index],list_module[max_index])
print('Min module:', list_sophuc[min_index],list_module[min_index])
```

```python
# 2. Phân số
import random
class Phanso:
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    def print(self):
        vt1 = ''
        vt2 = ''
        if self.a%self.b == 0:
            self.a = self.a/self.b
            self.b = 1
        if self.b == 0:
            return '#N/A'
        elif self.a == 0:
            return '0'
        elif self.b == 1:
            return str(self.a)
        elif self.b == -1:
            return str(-self.a)
        else:
            if self.b < 0:
                vt1 = str(-self.a)
                vt2 = str(-self.b)
            else:
                vt1 = str(self.a)
                vt2 = str(self.b)
            return vt1 + '/' + vt2
    
    def sosanh(self, other):
        if (self.a/self.b) > (other.a/other.b):
            return self.print() + ' > ' + other.print()
        elif (self.a/self.b) == (other.a/other.b):
            return self.print() + ' =' + other.print()
        else:
            return self.print() + ' < ' + other.print()
        
    def tong(self, other):
        return Phanso((self.a*other.b + other.a*self.b), (self.b*other.b)).print()
    
    def hieu(self, other):
        return Phanso((self.a*other.b - other.a*self.b), (self.b*other.b)).print()
    
    def tich(self, other):
        return Phanso((self.a*other.a), (self.b*other.b)).print()
    
    def thuong(self, other):
        return Phanso((self.a/other.a), (self.b/other.b)).print()
    
    def value(self):
        return self.a/self.b
    
ps1 = Phanso(3,-4)
print('PS1:',ps1.print())
ps2 = Phanso(14,7)
print('PS2:',ps2.print())
print('So sánh:', ps1.sosanh(ps2))
print('Tổng:', ps1.tong(ps2))
print('Hiệu:', ps1.hieu(ps2))
print('Tích:', ps1.tich(ps2))
print('Thương:', ps1.thuong(ps2))
print()

list_phanso = []
list_value = []
i = 0 
dict_ps = {}
range_int = [i for i in range(-10, 0)] + [j for j in range(1, 10)]
for i in range(20):
    temp = Phanso(random.choice(range_int), random.choice(range_int))
    list_phanso.append(temp.print())
    list_value.append(temp.value())
    dict_ps[temp.value()] = temp.print()
print(list_phanso)

max = list_value[0]
min = list_value[1]
max_index = 0
min_index = 1

for i in range(len(list_value)):
    if list_value[i] > max:
        max = list_value[i]
        max_index = i    
    if list_value[i] < min:
        min = list_value[i]
        min_index = i
print('Max:',list_phanso[max_index])
print('Min:',list_phanso[min_index])

list_ps = []
for keys, values in dict_ps.items():
    temp = []
    temp.append(keys)
    temp.append(values)
    list_ps.append(temp)
list_ps.sort()
list_ps2 = []
for i in list_ps:
    list_ps2.append(i[1])
print(list_ps2)
```

```python
# 3. List student
class Student:
    def __init__(self, msv, name, dob, add):
        self.msv = msv
        self.name = name
        self.dob = dob
        self.add = add
        
    def infor(self):
        list_std = []
        infor = []
        infor.append(self.name)
        infor.append(self.dob)
        infor.append(self.add)
        list_std.append(self.msv)
        list_std.append(infor)
        return list_std
    
std1 = Student(101, 'Mạnh', 1993, 'Hà Nội')
std2 = Student(102, 'Ánh', 1994, 'Bắc Giang')
print(std1.infor())
print()

class ListStudent:
    def __init__(self, *std):
        self.data = [*std]
        
    def print(self):
        list_sv = []
        for i in self.data:
            list_sv.append(i)
        return list_sv
    
    def add(self, std):
        self.data.append(std)
        
    def delete(self, std):
        self.data.remove(std)
        
    def update(self, msv, name, dob, add):
        for i in self.data:
            if i[0] == msv:
                i[1][0] = name
                i[1][1] = dob
                i[1][2] = add
    
    def find(self, msv):
        for i in self.data:
            if i[0] == msv:
                return i
            
ListStd1 =  ListStudent(std1.infor(), std2.infor())
print('In ra danh sách:')
print(ListStd1.print())
print()

std3 = Student(103, 'Hường', 1995, 'Bắc Ninh')
ListStd1.add(std3.infor())
print('Thêm người vào danh sách:')
print(ListStd1.print())
print()

ListStd1.delete(std2.infor())
print('Xóa người khỏi danh sách:')
print(ListStd1.print())
print()

ListStd1.update(103, 'Huyền', 1996, 'Bắc Giang')
print('Update danh sách theo MSV:')
print(ListStd1.print())
print()

print('Tìm thông tin theo MSV:')
print(ListStd1.find(103))
                
```

```python
# 2.2 Phân số
class PS:
    def __init__(self, ts, ms):
        self.ts = ts
        if ms!=0:
            self.ms = ms
    
    def output(self):
        print('{0}/{1}'.format(self.ts, self.ms))
        
    def compare(self, other):
        if self.ts*other.ms > self.ms*other.ts:
            return True
        else:
            return False
        
    def sum(self, other):
        _ts = self.ts*other.ms + self.ms*other.ts
        _ms = self.ms*other.ms
        return PS(_ts,_ms)
    
    def copy(self):
        return PS(self.ts, self.ms)
        
ps = PS(2,6)
ps2 = PS(1,2)
ps.output()
print(ps.compare(ps2))
ps.sum(ps2).output()
```

```python
class Student2:
    def __init__(self, code, fullname, birthday, residence):
        self.code = code
        self.fullname = fullname
        self.birthday = birthday
        self.residence = residence
        
    def infor(self):
        print(f'{self.code}\t\t{self.fullname}\t\t{self.birthday}\t\t{self.residence}')
```

```python
class ListStd:
    def __init__(self, _list):
        self.list = _list
        
    def initialize(self):
        st1 = Student2('220033','Phạm Thu Hoài','22/10/1998','Hà Nội')
        st2 = Student2('220034','Thanh Huyền','22/10/1998','Bắc Ninh')
        st3 = Student2('220035','Trần Công Hoan','22/10/1998','Hưng Yên')
        st4 = Student2('220036','Lã Văn Mạnh','22/10/1998','Hải Phòng')
        st5 = Student2('220037','Phạm Ngọc Bình','22/10/1998','Hà Nam')
        self.list = [st1, st2, st3, st4, st5]
    
    def output(self):
        print('MSV\t\tHọ tên\t\t\tNgày sinh\t\tQuê quán')
        for i in self.list:
            i.infor()
    
    def add(self, std):
        self.list.append(std)
        
    def update(self, code, std):
        '''code: Mã sv sẽ bị chỉnh sửa
        std: thông tin mới sv sẽ được cập nhật'''
        for i in self.list:
            if i.code == code:
                i.fullname = std.fullname
                i.birthday = std.birthday
                i.residence = std.residence
                
    def delete(self, code):
        for i in self.list:
            if i.code == code:
                self.list.remove(i)
                break
    
    def search(self, code):
        for i in self.list:
            if i.code == code:
                return i
```

```python
class StudentManagement:
    list_st = None
    def menu(self):
        print('---------QUẢN LÝ SINH VIÊN---------')
        print('- 0. Tải danh mục sinh viên      --')
        print('- 1. Hiển thị danh mục sinh viên --')
        print('- 2. Thêm sinh viên              --')
        print('- 3. Cập nhật sinh viên          --')
        print('- 4. Xóa sinh viên               --')
        print('- 5. Tìm kiếm sinh viên          --')
        print('- 6. Thoát                       --')
        print('-----------------------------------')
        
    def run(self):
        while True:
            self.menu()
            action = input('Nhập vào lựa chọn: ')
            if action == '0':
                self.list_st = ListStd(None)
                self.list_st.initialize()
                print('Tải thành công')
                input('Bấm Enter để tiếp tục')
                
            elif action == '1':
                if self.list_st == None:
                    print('Chưa tải danh mục')    
                else:
                    self.list_st.output()
                input('Bấm Enter để tiếp tục')
                
            elif action == '2':
                code =  input('Nhập mã sinh viên: ')
                fullname = input('Nhập họ tên: ')
                birthday = input('Nhập ngày sinh: ')
                residence = input('Nhập quê quán: ')
                adding_student = Student2(code, fullname, birthday, residence)
                self.list_st.add(adding_student)
                print('Thêm sinh viên thành công')
                input('Bấm Enter để tiếp tục')
                
            elif action == '3':
                pass
            elif action == '4':
                pass
            elif action == '5':
                pass    
            elif action == '6':
                break
        print("Chương trình kết thúc")
```
