```python
from tkinter import *

root = Tk()
root.title('Calculator')
root.geometry('305x390')
root.resizable(0,0)

screen = Entry(root, font=('Helvetica',17), justify='right')
screen.grid(row=0, column=0, columnspan=5, padx=5, pady=5, ipadx=10, ipady=10)
screen.insert(0,'0')

text_display = ''
first_number=0
second_number=0
action=''

def number_press(event):
    global text_display
    pressed = event.widget['text']
    if text_display == '' or text_display == '0': 
        if pressed=='0':
            text_display = '0'
        else:
            text_display =''
            text_display += pressed
    else:
        text_display += pressed
    screen.delete(0,'end')
    screen.insert(0, text_display)
    
def action_press(event):
    global first_number
    first_number = float(screen.get())
    global action
    action = event.widget['text']
    global text_display
    text_display = ''
    
def finish_press(event):
    global second_number
    second_number = float(screen.get())
    if action=='+':
        result = first_number+second_number
    elif action=='-':
        result = first_number-second_number
    elif action=='*':
        result = first_number*second_number
    elif action=='/':
        if second_number == 0:
            result = 'DIV/0'
        else:
            result = first_number/second_number
    if str(result)[-2:] == '.0':
        result = int(result)
    screen.delete(0,'end')
    screen.insert(0, result)

def clear_press(event):
    global text_display
    global first_number
    global second_number
    global action
    text_display = ''
    first_number=0
    second_number=0
    action=''
    screen.delete(0,'end')
    screen.insert(0,'0')
    
def self_press(event):
    temp = float(screen.get())
    pressed = event.widget['text'] 
    if pressed == '1/x':
        if temp != 0:
            result = 1/temp
        else:
            result = 'NaN'
    elif pressed == '%':
        result = temp/100    
    elif pressed == '+/-':
        result = -temp
    elif pressed == 'sqrt':
        if temp >= 0:
            result = temp**0.5
        else:
            result = 'NaN'
    global text_display
    text_display = str(result)
    screen.delete(0,'end')
    screen.insert(0, result)
    
def dot_press(event):
    temp = screen.get()
    if '.' not in temp:
        temp += '.'
    global text_display
    text_display = temp
    screen.delete(0,'end')
    screen.insert(0, temp)
    
keyboards = {'MC': 0, 'MR': 0, 'MS': 0, 'M+': 0, 'M-': 0,
             '<-': 0, 'CE': 0, 'C': 4, '+/-': 5, 'sqrt': 5, 
             '7': 1, '8': 1,'9': 1, '/': 2, '%': 5, 
             '4': 1, '5': 1, '6': 1, '*': 2, '1/x': 5, 
             '1': 1, '2': 1, '3': 1, '-': 2, '=': 3, 
             '0': 1, '': 0, '.': 6,  '+': 2}

keys=list(keyboards.keys())
row = 1
column = 0

for i in range(len(keys)):
    if i%5==0 and i!=0:
        row+=1
    # phÃ­m 0
    if i==26:
        continue
    if i==25:
        btn=Button(root, text=keys[i], font=('Helvetica',10, 'bold'), width=13, height=2)
        btn.grid(row=row,column=i%5, padx=5, pady=5, columnspan=2)
    elif i==24:
        btn=Button(root, text=keys[i], font=('Helvetica',10, 'bold'), width=5, height=5)
        btn.grid(row=row,column=i%5, padx=5, pady=5, rowspan=2, ipady=3)
    else:
        btn = Button(root, text=keys[i], font=('Helvetica',10, 'bold'), width=5, height=2)
        btn.grid(row=row,column=i%5, padx=5, pady=5)
    if keyboards[keys[i]]==1:
        btn.bind('<Button-1>', number_press)
    elif keyboards[keys[i]]==2:
        btn.bind('<Button-1>', action_press)
    elif keyboards[keys[i]]==3:
        btn.bind('<Button-1>', finish_press)
    elif keyboards[keys[i]]==4:
        btn.bind('<Button-1>', clear_press)
    elif keyboards[keys[i]]==5:
        btn.bind('<Button-1>', self_press)
    elif keyboards[keys[i]]==6:
        btn.bind('<Button-1>', dot_press)
root.mainloop()
```
