```python
# khai báo thư viện, điều chỉnh thông số
import turtle
tl = turtle.Turtle()
tl.shape('turtle')

screen = tl.getscreen()
screen.bgcolor('black')

tl.speed(30)

tl.fillcolor('white')
tl.begin_fill()
tl.circle(100)
tl.end_fill()

tl.penup()
tl.left(90)
tl.forward(170)
tl.right(90)
tl.pendown()
tl.fillcolor('white')
tl.begin_fill()
tl.circle(70)
tl.end_fill()

tl.penup()
tl.right(90)
tl.forward(40)
tl.pendown()
tl.fillcolor('black')
tl.begin_fill()
tl.circle(5)
tl.end_fill()

tl.penup()
tl.forward(15)
tl.pendown()
tl.fillcolor('black')
tl.begin_fill()
tl.circle(5)
tl.end_fill()

tl.penup()
tl.forward(15)
tl.pendown()
tl.fillcolor('black')
tl.begin_fill()
tl.circle(5)
tl.end_fill()

tl.penup()
tl.backward(65)
tl.left(90)
tl.pendown()
tl.fillcolor('red')
tl.begin_fill()
tl.forward(60)
tl.circle(12, 180)
tl.forward(120)
tl.circle(12, 180)
tl.forward(60)
tl.end_fill()

tl.penup()
tl.forward(25)
tl.pendown()
tl.right(90)
tl.fillcolor('red')
tl.begin_fill()
tl.forward(45)
tl.circle(5,180)
tl.forward(45)
tl.end_fill()

tl.penup()
tl.right(90)
tl.forward(5)
tl.right(90)
tl.pendown()
tl.fillcolor('red')
tl.begin_fill()
tl.forward(65)
tl.circle(10,180)
tl.forward(65)
tl.end_fill()

tl.penup()
tl.left(90)
tl.forward(25+10+5+20)
tl.right(90)
tl.forward(45)
tl.right(90)
tl.pendown()
tl.pensize(5)
tl.fillcolor('black')
tl.begin_fill()
tl.circle(70, 30)
tl.circle(70, -60)
tl.circle(70, 30)
tl.end_fill()

tl.penup()
tl.left(90)
tl.forward(20)
tl.right(90)
tl.pendown()
tl.pensize(1)
tl.fillcolor('red')
tl.begin_fill()
tl.circle(10)
tl.end_fill()

tl.penup()
tl.forward(30)
tl.left(90)
tl.forward(30)
tl.right(90)
tl.fillcolor('black')
tl.begin_fill()
tl.circle(5)
tl.end_fill()

tl.penup()
tl.backward(60)
tl.pendown()
tl.fillcolor('black')
tl.begin_fill()
tl.circle(5)
tl.end_fill()


tl.penup()
tl.backward(70)
tl.left(80)
tl.forward(50)
tl.right(90)
tl.forward(25)
tl.pendown()
tl.fillcolor('grey')
tl.begin_fill()
tl.forward(125)
tl.left(90)
tl.forward(20)
tl.left(90)
tl.forward(100)
tl.left(90)
tl.forward(20)
tl.end_fill()


turtle.done()
```
