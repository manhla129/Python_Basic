```python
#1 Vẽ hình sao theo yêu cầu
n = int(input('n = '))
m = int(input('m = '))
if n%2==0:
    n-=1
for i in range(2*n):
    if i <= n:
        if i%2!=0:
            print((' '*int((n-i)/2) + '*'*i+ ' '*int((n-i+2)/2))*m)
    else:
        if i%2!=0:
            print((' '*int((i-n)/2)+'*'*(2*n-i)+' '*int((i-n+2)/2))*m)

```

```python
#2 Số cách cộng tiền 2, 3, 5k thành tổng 100k
step = 0
for i in range(21):
    for j in range(34):
        for k in range(51):
            if i*5 + j*3 + k*2==100:
                step+=1
                print(f'Cách {step}:')
                print('5*{0} + 3*{1} + 2*{2} = {3} + {4} + {5} = 100'.format(i,j,k,5*i,3*j,2*k))
print(f'Tổng số cách: {step}')
```
