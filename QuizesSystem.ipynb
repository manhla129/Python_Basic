```python
class Answer:
    def __init__(self, text, iscorrect):
        self.text = text
        self.iscorrect = iscorrect
        
    def display(self):
        print("Đáp án:", self.text, end=' ')
        print(self.iscorrect)
        
class Question:
    def __init__(self, text, ans):
        self.text = text
        self.ans = ans
        
   
    
    def display(self):
        print("Nội dung câu hỏi:", self.text)
        for i in self.ans:
            i.display()
            
class Test:
    def __init__(self, test):
        self.test = test
    
    def createThreeQuestion(self):
        ans1 = Answer('Việt Nam',True)
        ans2 = Answer('Mexico',False)
        ans3 = Answer('Pháp',False)
        ans4 = Answer('Úc',False)
        ans01 = [ans1, ans2, ans3, ans4]         
        q1 = Question('Nước nào thuộc Châu Á?', ans01)

        ans5 = Answer('Mexico',False)
        ans6 = Answer('Việt Nam',True)
        ans7 = Answer('Pháp',False)
        ans8 = Answer('Úc',False)
        ans02 = [ans5, ans6, ans7, ans8]         
        q2 = Question('Hà Nội là thủ đô nước nào?', ans02)

        ans9 = Answer('Mexico',False)
        ans10 = Answer('Việt Nam',False)
        ans11 = Answer('Trung Quốc',False)
        ans12 = Answer('Nga',True)
        ans03 = [ans9, ans10, ans11, ans12]         
        q3 = Question('Nước nào rộng nhất thế giới?', ans03)
        self.test = [q1, q2, q3]
      
```

```python
from tkinter import *

index = 0
t = Test(None)
t.createThreeQuestion()

def displayQuestion(i):
    question_text.delete(1.0,'end')
    question_text.insert(1.0,t.test[i].text)
    
    text_answer1.delete(1.0,'end')
    text_answer1.insert(1.0,t.test[i].ans[0].text)
    text_answer2.delete(1.0,'end')
    text_answer2.insert(1.0,t.test[i].ans[1].text)
    text_answer3.delete(1.0,'end')
    text_answer3.insert(1.0,t.test[i].ans[2].text)
    text_answer4.delete(1.0,'end')
    text_answer4.insert(1.0,t.test[i].ans[3].text)

root = Tk()
root.geometry('750x600')
root.resizable(0,0)
root.title('Question')

lbFrame_question = LabelFrame(root, text='Nội dung câu hỏi',font=('Helvetica',12), fg='blue')
lbFrame_question.pack(expand=True, fill=BOTH)
question_text = Text(lbFrame_question, font=('Helvetica',14), height=5)
question_text.pack(expand=True, fill=BOTH, padx=10, pady=10)


lbFrame_answers = LabelFrame(root, text='Các đáp án',font=('Helvetica',12), fg='blue')
lbFrame_answers.pack(expand=True, fill=BOTH)
cb1 = BooleanVar()
cb1.set(0)
cb2 = BooleanVar()
cb2.set(0)
cb3 = BooleanVar()
cb3.set(0)
cb4 = BooleanVar()
cb4.set(0)

ckbtn1 = Checkbutton(lbFrame_answers, variable=cb1, text='', onvalue=True, offvalue=False)
ckbtn1.grid(row=0, column=0, padx=5, pady=5)
ckbtn2 = Checkbutton(lbFrame_answers, variable=cb2, text='', onvalue=True, offvalue=False)
ckbtn2.grid(row=0, column=2, padx=5, pady=5)
ckbtn3 = Checkbutton(lbFrame_answers, variable=cb3, text='', onvalue=True, offvalue=False)
ckbtn3.grid(row=1, column=0, padx=5, pady=5)
ckbtn4 = Checkbutton(lbFrame_answers, variable=cb4, text='', onvalue=True, offvalue=False)
ckbtn4.grid(row=1, column=2, padx=5, pady=5)

text_answer1 = Text(lbFrame_answers, font=('Helvetica',12), height=5, width=35)
text_answer1.grid(row=0, column=1, padx=5, pady=5)
text_answer2 = Text(lbFrame_answers, font=('Helvetica',12), height=5, width=35)
text_answer2.grid(row=0, column=3, padx=5, pady=5)
text_answer3 = Text(lbFrame_answers, font=('Helvetica',12), height=5, width=35)
text_answer3.grid(row=1, column=1, padx=5, pady=5)
text_answer4 = Text(lbFrame_answers, font=('Helvetica',12), height=5, width=35)
text_answer4.grid(row=1, column=3, padx=5, pady=5)

lbFrame_action = LabelFrame(root, text='Thao tác',font=('Helvetica',14), fg='blue')
lbFrame_action.pack(expand=True, fill=BOTH)

photo_previous = PhotoImage(file=r'IMG\previous.png')
btn_previous = Button(lbFrame_action, text='Câu hỏi trước',font=('Helvetica',8),
                     image=photo_previous, compound=LEFT)
btn_previous.place(x=10, y=10)

photo_next = PhotoImage(file=r'IMG\next.png')
btn_next = Button(lbFrame_action, text='Câu hỏi tiếp theo',font=('Helvetica',8),
                     image=photo_next, compound=RIGHT)
btn_next.place(x=140, y=10)

photo_check = PhotoImage(file=r'IMG\check.png')
btn_end = Button(lbFrame_action, text='Kết thúc',font=('Helvetica',8), fg='red',
                     image=photo_check, compound=LEFT)
btn_end.place(x=670, y=10)
displayQuestion(0)

root.mainloop()
```
