```python
# Hàm phân loại theo điểm
def xeploai(mark):
    '''phân loại học lực theo điểm
    0-5: Kém
    5-6: TB
    6-8: Khá
    8-10: Giỏi
    '''
    if mark < 5:
        return 'Kém'
    elif mark <= 6:
        return 'TB'
    elif mark <= 8:
        return 'Khá'
    elif mark <= 10:
        return 'Giỏi'
    else:
        return '#N/A'

# Nhập danh sách sinh viên
dict_std = {}
n = int(input('Nhập số sinh viên: '))
for i in range(1, n+1):
    gen = input('Nhập MSV của SV {0}'.format(i))
    name = input('Nhập tên của SV {0}'.format(i))
    mark = input('Nhập điểm của SV {0}'.format(i))
    infor = (name, mark)
    dict_std[gen] = infor
print(dict_std)

# 1. In ra danh sách sinh viên có điểm >=8
list_8 = []
for keys, values in dict_std.items():
    if float(values[1])>=8:
        list_8.append(values[0])
print('1. Danh sách sinh viên có điểm >=8:', list_8)

# 2. Xếp loại sinh viên theo điểm
print('2. Xếp loại sinh viên theo điểm')
print('\tMSV\t\tHọ tên\t\tĐiểm\t\tXếp loại')
for keys, values in dict_std.items():
    print('\t{0}'.format(keys), end='\t\t')
    print(values[0], end='\t\t')
    print(values[1], end='\t\t')
    print(xeploai(float(values[1])))
    
# 3. In danh sách sinh viên sắp xếp theo tên
list_std = []
for keys, values in dict_std.items():
    list_temp = []
    list_temp.append(values[0])
    list_temp.append(keys)
    list_temp.append(values[1])
    list_std.append(list_temp)
list_std.sort()

dict_std_new = {}
for i in list_std:
    gen = i[1]
    name = i[0]
    mark = i[2]
    infor = (name, mark)
    dict_std_new[gen] = infor

print('3. Danh sách đã sắp xếp theo tên')
print('\tMSV\t\tHọ tên\t\tĐiểm')
for keys, values in dict_std_new.items():
    print('\t{0}'.format(keys), end='\t\t')
    print(values[0], end='\t\t')
    print(values[1])
```
